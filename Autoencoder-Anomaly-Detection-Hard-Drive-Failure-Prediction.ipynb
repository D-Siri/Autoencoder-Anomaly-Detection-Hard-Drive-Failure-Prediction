```python
from sklearn.model_selection import train_test_split

normal_instances = df[df['failure'] == 0]
anomalous_instances = df[df['failure'] == 1]

normal_train, normal_test = train_test_split(normal_instances, test_size=0.2, random_state=42)
normal_train, normal_val = train_test_split(normal_train, test_size=0.25, random_state=42)  # 0.25 x 0.8 = 0.2

X_test = pd.concat([normal_test, anomalous_instances])

y_train = normal_train['failure']
y_val = normal_val['failure']
y_test = X_test['failure']

X_train = normal_train.drop(columns=['failure'])
X_val = normal_val.drop(columns=['failure'])
X_test = X_test.drop(columns=['failure'])
```

```python
from keras.layers import Input, Dense, LeakyReLU
from keras.models import Model
def autoencoder_1(input_dim):
    input_layer = Input(shape=input_dim)
    encoded = Dense(128, activation='relu')(input_layer)
    encoded = Dense(64, activation='relu')(encoded)
    encoded = Dense(32, activation='relu')(encoded)
    
    decoded = Dense(64, activation='relu')(encoded)
    decoded = Dense(128, activation='relu')(decoded)
    decoded = Dense(input_dim, activation='linear')(decoded)
    
    autoencoder = Model(input_layer, decoded)
    autoencoder.compile(optimizer='adam', loss='mse')
    return autoencoder

def autoencoder_2(input_dim):
    input_layer = Input(shape=input_dim)
    encoded = Dense(256)(input_layer)
    encoded = LeakyReLU(alpha=0.2)(encoded)
    encoded = Dense(128)(encoded)
    encoded = LeakyReLU(alpha=0.2)(encoded)
    encoded = Dense(64)(encoded)
    encoded = LeakyReLU(alpha=0.2)(encoded)
    
    decoded = Dense(128)(encoded)
    decoded = LeakyReLU(alpha=0.2)(decoded)
    decoded = Dense(256)(decoded)
    decoded = LeakyReLU(alpha=0.2)(decoded)
    decoded = Dense(input_dim, activation='linear')(decoded)
    
    autoencoder = Model(input_layer, decoded)
    autoencoder.compile(optimizer='adam', loss='mse')
    return autoencoder

def autoencoder_3(input_dim):
    input_layer = Input(shape=(input_dim,))
    encoded = Dense(32, activation='relu')(input_layer)
    encoded = Dense(16, activation='relu')(encoded)
    decoded = Dense(input_dim, activation='linear')(encoded)
    
    autoencoder = Model(inputs=input_layer, outputs=decoded)
    autoencoder.compile(optimizer='adam', loss='mean_squared_error')
    
    return autoencoder

input_dim = X_train.shape[1]  
autoencoder1 = autoencoder_1(input_dim)
autoencoder1.summary()

autoencoder2 = autoencoder_2(input_dim)
autoencoder2.summary()

autoencoder3 = autoencoder_3(input_dim)
autoencoder3.summary()
```

```python
from keras.callbacks import Callback
from keras.callbacks import EarlyStopping, ModelCheckpoint
import numpy as np

early_stopping = EarlyStopping(monitor='val_loss', patience=3, restore_best_weights=True)
checkpoint = ModelCheckpoint('best_autoencoder_model.h5', monitor='val_loss', save_best_only=True)

class AccuracyHistory(Callback):
    def __init__(self):
        super().__init__()
        self.acc_history = {'accuracy': [], 'val_accuracy': []}

    def calculate_reconstruction_accuracy(self, X):
        X_pred = self.model.predict(X)
        mse = np.mean(np.power(X - X_pred, 2), axis=1)
        max_possible_loss = np.mean(np.power(X, 2), axis=1)  # Maximum possible MSE
        reconstruction_accuracy = 1 - mse / max_possible_loss
        return reconstruction_accuracy

    def on_epoch_end(self, epoch, logs=None):
        if logs is not None:
            train_accuracy = self.calculate_reconstruction_accuracy(X_train)
            val_accuracy = self.calculate_reconstruction_accuracy(X_val)
            mean_train_accuracy = np.mean(train_accuracy)
            mean_val_accuracy = np.mean(val_accuracy)
            self.acc_history['accuracy'].append(mean_train_accuracy)
            self.acc_history['val_accuracy'].append(mean_val_accuracy)
            print(f'Epoch {epoch + 1} - Training Accuracy: {mean_train_accuracy}, Validation Accuracy: {mean_val_accuracy}')

accuracy_history1 = AccuracyHistory()
accuracy_history2 = AccuracyHistory()
accuracy_history3 = AccuracyHistory()
history1 = autoencoder1.fit(X_train, X_train,
                            epochs=50,
                            batch_size=32,
                            shuffle=True,
                            validation_data=(X_val, X_val),
                            callbacks=[early_stopping, checkpoint, accuracy_history1])

print(accuracy_history1.acc_history['accuracy'])
print(accuracy_history1.acc_history['val_accuracy'])
```

```python

early_stopping = EarlyStopping(monitor='val_loss', patience=3, restore_best_weights=True)
checkpoint = ModelCheckpoint('best_autoencoder_model2.h5', monitor='val_loss', save_best_only=True)

history2 = autoencoder2.fit(X_train, X_train,
                            epochs=50,
                            batch_size=32,
                            shuffle=True,
                            validation_data=(X_val, X_val),
                            callbacks=[early_stopping, checkpoint, accuracy_history2])

print(accuracy_history2.acc_history['accuracy'])
print(accuracy_history2.acc_history['val_accuracy'])

```

```python

early_stopping = EarlyStopping(monitor='val_loss', patience=3, restore_best_weights=True)
checkpoint = ModelCheckpoint('best_autoencoder_model3.h5', monitor='val_loss', save_best_only=True)


history3 = autoencoder3.fit(X_train, X_train,
                            epochs=50,
                            batch_size=32,
                            shuffle=True,
                            validation_data=(X_val, X_val),
                            callbacks=[early_stopping, checkpoint, accuracy_history3])

print(accuracy_history3.acc_history['accuracy'])
print(accuracy_history3.acc_history['val_accuracy'])
```

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_loss(history, model_name):
    plt.plot(history.history['loss'], label='Training Loss')
    plt.plot(history.history['val_loss'], label='Validation Loss')
    plt.title(f'{model_name} Training and Validation Loss')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.legend()
    plt.show()

plot_loss(history1, 'Autoencoder 1')
plot_loss(history2, 'Autoencoder 2')
plot_loss(history3, 'Autoencoder 3')
```

```python
import matplotlib.pyplot as plt

def plot_accuracy(accuracy_history, name):
    plt.plot(range(1, len(accuracy_history['accuracy']) + 1), accuracy_history['accuracy'], label='Training Accuracy')
    plt.plot(range(1, len(accuracy_history['val_accuracy']) + 1), accuracy_history['val_accuracy'], label='Validation Accuracy')
    plt.xlabel('Epochs')
    plt.ylabel('Accuracy')
    plt.title(f'Training and Validation Accuracy ({name})')
    plt.legend()
    plt.grid(True)
    plt.show()

plot_accuracy(accuracy_history1.acc_history, 'Autoencoder 1')
plot_accuracy(accuracy_history2.acc_history, 'Autoencoder 2')
plot_accuracy(accuracy_history3.acc_history, 'Autoencoder 3')
```

```python
def get_test_accuracy(autoencoder_model, X_test, y_test, percentile=95):
   
    X_pred = autoencoder_model.predict(X_test)
    
    reconstruction_errors = np.mean(np.square(X_test - X_pred), axis=1)
    
    threshold = np.percentile(reconstruction_errors, percentile)
    
    y_pred = (reconstruction_errors > threshold).astype(int)
    
    accuracy = (y_pred == y_test).mean()
    
    return accuracy

accuracy1 = get_test_accuracy(autoencoder1, X_test, y_test)
accuracy2 = get_test_accuracy(autoencoder2, X_test, y_test)
accuracy3 = get_test_accuracy(autoencoder3, X_test, y_test)
print("Test Accuracy for Autoencoder 1:", accuracy1)
print("Test Accuracy for Autoencoder 2:", accuracy2)
print("Test Accuracy for Autoencoder 3:", accuracy3)
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score

def evaluation(reconstruction_errors, y_test, name):
    top_percentile = 5  
    threshold_index = int(len(reconstruction_errors) * top_percentile / 100)
    threshold = np.sort(reconstruction_errors)[-threshold_index]

    y_pred = (reconstruction_errors > threshold).astype(int)

    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    
    print(f"Evaluation metrics for {name}:")
    print("Precision:", precision)
    print("Recall:", recall)
    print("F1 Score:", f1)

evaluation(reconstruction_error1, y_test, "Autoencoder 1")
evaluation(reconstruction_error2, y_test, "Autoencoder 2")  
evaluation(reconstruction_error3, y_test, "Autoencoder 3")
```
